```python

import matplotlib.pyplot as plt

t = []
y = []
with open("train.dat") as f:
    lines = f.readlines()
    for i in lines[1:]:
      arr = i.split()
      t.append(float(arr[0]))
      y.append(float(arr[1]))
plt.plot(t,y)
```

```python
import numpy as np

# массив локальных минимумов и максимумов
arrDy=np.sign(np.diff(y))

# массив индексов максимумов, где 2 это минимум, а -2 максимум
indexesEx=np.nonzero(np.diff(arrDy)==-2)


#Рисую на графике максимумы
plt.plot(t,y)
for i in indexesEx[0]:
  plt.scatter(t[i], y[i], color='orange', s=40, marker='o')

#Находим количество промежутков между максимумами, 
#что по сути и является количеством гармонических составляющих.
countEx = len(indexesEx[0]) -1
countEx

#На самом деле мне кажется, что правильней это было сделать через 
#преобразование Фурье, чтобы определить гармонику сигнала, но у меня не получилось :c

```
